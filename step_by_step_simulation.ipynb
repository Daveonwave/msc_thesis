#### With random demand pattern (24h)

```python
import random

new_pattern = []
for i in range(0, 24):
    new_pattern.append(round(random.uniform(0.5,1.5), 2))

new_pattern = pd.Series(new_pattern)
new_pattern
```

```python
net2 = network.WaterDistributionNetwork("anytown.inp")
net2.set_time_params(duration=86400, hydraulic_step=600) # duration=24h, hydstep=10min
net2.set_basedemand_pattern('2', new_pattern)
```

```python
net2.run()
```

```python
net2.df_junctions_report.sample(10).sort_index()
```
